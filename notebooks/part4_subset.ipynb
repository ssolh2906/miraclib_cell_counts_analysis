```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../cell_counts.db")

# Independent check of Part 4 headline number, hand-written SQL (not reusing src/ code).
query = """
SELECT cc.cell_count
FROM cell_counts cc
JOIN samples sm ON cc.sample_id = sm.sample_id
JOIN subjects subj ON sm.subject_id = subj.subject_id
WHERE subj.condition = 'melanoma' AND subj.treatment = 'miraclib' AND sm.sample_type = 'PBMC'
  AND sm.time_from_treatment_start = 0 AND subj.sex = 'M' AND subj.response = 'yes'
  AND cc.population = 'b_cell'
"""
b_cells = pd.read_sql(query, conn)
len(b_cells), round(b_cells["cell_count"].mean(), 2)
```
